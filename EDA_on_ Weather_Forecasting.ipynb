## Univariate Analysis

```python
numerical_data["Feels_Like °C"]
```

```python
numerical_data["Feels_Like °C"].describe()
```

```python
plt.hist(numerical_data["Feels_Like °C"],bins=12)
plt.xlabel("Feels_Like °C")
plt.ylabel("Frequency")
plt.title("Histogram on Feels_Like °C")
plt.grid()
plt.show()
```

## Observation
- The Feels_Like temperature mostly ranges between 28°C and 32°C.
- The most frequent (mode) values are around 30°C and 31°C, each appearing nearly 100 times, indicating these are the most common comfort levels recorded.
- Fewer readings are observed at 33°C, showing that very hot perceived temperatures are less common.
- The occurrences drop slightly below 28°C, meaning lower temperature perceptions are also less frequent.

```python
sns.histplot(numerical_data["Feels_Like °C"], bins=12, kde=True)
plt.title("Histogram on Feels_Like °C using seaborn")
plt.show()
```

```python
plt.hist(numerical_data["Wind_Speed_km/h"],bins=12)
plt.xlabel("Wind_Speed_km/h")
plt.ylabel("Frequency")
plt.title("Histogram on Wind_Speed_km/h")
plt.grid()
plt.show()
```

## Observation
- The most frequent wind speed is around 7–8 km/h, with a peak frequency of roughly 70+ occurrences — suggesting that light to moderate breezes are the most common.
- Wind speeds between 2–5 km/h and 10–16 km/h are less frequent, but still occur consistently, indicating some variability in wind conditions.
- The distribution is slightly bimodal — one peak around 7–8 km/h and a smaller cluster near 14–16 km/h, showing occasional higher wind speeds.

```python
sns.histplot(numerical_data["Wind_Speed_km/h"], bins=12, kde=True)
plt.title("Histogram on Wind_Speed_km/h using seaborn")
plt.show()
```

## Multivariate Analysis

###  City ,Weather_Condition,Humidity%

```python
table = df.pivot_table(values='Humidity%', index='Weather_Condition', columns='City', aggfunc='count')
```

```python
sns.heatmap(table, annot=True, cmap='coolwarm')
plt.title('Humidity by City and Weather Condition')
plt.show()
```

```python
table = df.pivot_table(values='City', index='Rain', columns='Weather_Condition', aggfunc='count')
```

```python
sns.heatmap(table, annot=True, cmap='coolwarm')
plt.title('Rain by City and Weather Condition')
plt.show()
```

```python
table = df.pivot_table(values='Feels_Like °C', index='Weather_Condition', columns='City', aggfunc='mean')
```

```python
sns.heatmap(table, annot=True, cmap='coolwarm')
plt.title('Feels_Like °C by City and Weather Condition')
plt.show()
```
