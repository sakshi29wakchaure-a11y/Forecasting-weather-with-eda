```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
import re
```

```python
df = pd.read_csv('WetherForcasting.csv')
df
```

```python
df=df.drop(columns='Wind Direction')
df
```

```python
df.rename(columns={'Feels Like':'Feels_Like °C'},inplace=True)
df
```

```python
df.rename(columns={'Wind Speed':'Wind_Speed_km/h'},inplace=True)
df
```

```python
df['Feels_Like °C'] = pd.to_numeric(df['Feels_Like °C'].astype(str).str.replace(r"[^0-9\-]", "", regex=True),errors='coerce').astype('Int64')
df
```

```python
df['Wind_Speed_km/h'] = pd.to_numeric(df['Wind_Speed_km/h'].astype(str).str.replace(r"[^0-9\-]", "", regex=True),errors='coerce').astype('Int64')
df
```

```python
df.rename(columns={'Humidity':'Humidity%'},inplace=True)
df
```

```python
df.rename(columns={'Dew Point':'Dew_Point_%'},inplace=True)
df
```

```python
df['Dew_Point_%'] = (df['Dew_Point_%'].astype(str).str.replace('%', '', regex=False).str.strip())
df
```

```python
df['Humidity%'] = (df['Humidity%'].astype(str).str.replace('%', '', regex=False).str.strip())
df
```

```python
df.rename(columns={'Visibility': 'Visibility_mm'}, inplace=True)
df
```

```python
df['Visibility_mm'] = (df['Visibility_mm'].astype(str).str.replace('km', '', regex=False).str.replace('m', '', regex=False).str.strip())
df['Visibility_mm'] = pd.to_numeric(df['Visibility_mm'], errors='coerce')
df
```

```python
df.info()
```

```python
df.rename(columns={'Probability of Precipitation': 'Precip_Prob_Cat'}, inplace=True)
df
```

```python
df['Precip_Prob_Cat'] = (df['Precip_Prob_Cat'].astype(str).str.extract(r'(High|Moderate|Low)', flags=re.IGNORECASE)[0].str.title())
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df = df.drop(columns=['Date'])
df
```

```python
df = df.drop(columns=['Snow'])
df
```

```python
df.rename(columns={'Temperature': 'Weather_Condition'}, inplace=True)
df
```

```python
df['Weather_Condition'] = df['Weather_Condition'].astype(str).str.split('.').str[0].str.strip()
df
```
