```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
City = []
Date = []
Hour = []
Temperature = []
Feels_Like = []
Wind_Speed = []
Wind_Direction = []
Humidity = []
Dew_Point = []
Visibility = []
Precipitation_Prob = []
Rain = []
Snow = []


cities = {
    "Pune": "https://www.timeanddate.com/weather/india/pune/ext",
    "Mumbai": "https://www.timeanddate.com/weather/india/mumbai/ext",
    "Nashik": "https://www.timeanddate.com/weather/india/nashik/ext",
    "Nagpur": "https://www.timeanddate.com/weather/india/nagpur/ext"
}

headers = {"User-Agent": "Mozilla/5.0"}
max_rows = 400

for city_name, url in cities.items():
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")

   
    table = soup.find("table", class_="zebra tb-wt fw va-m tb-hover")
    if not table:
        print(f"No table found for {city_name}")
        continue

    rows = table.find("tbody").find_all("tr")

   
    for row in rows:
        cols = row.find_all("td")
        if len(cols) < 12:
            continue

        date_val = cols[0].text.strip() if cols[0] else "N/A"
        forecast_val = cols[1].text.strip() if cols[1] else "N/A"
        temp_val = cols[2].text.strip() if cols[2] else "N/A"
        feels_val = cols[3].text.strip() if cols[3] else "N/A"
        wind_speed_val = cols[4].text.strip() if cols[4] else "N/A"
        wind_dir_val = cols[5].text.strip() if cols[5] else "N/A"
        humidity_val = cols[6].text.strip() if cols[6] else "N/A"
        dew_val = cols[7].text.strip() if cols[7] else "N/A"
        visibility_val = cols[8].text.strip() if cols[8] else "N/A"
        precip_val = cols[9].text.strip() if cols[9] else "N/A"
        rain_val = cols[10].text.strip() if cols[10] else "N/A"
        snow_val = cols[11].text.strip() if cols[11] else "N/A"
        
        for hour in range(1, 25):
            City.append(city_name)
            Date.append(date_val)
            Hour.append(f"{hour}:00")
            Temperature.append(temp_val)
            Feels_Like.append(feels_val)
            Wind_Speed.append(wind_speed_val)
            Wind_Direction.append(wind_dir_val)
            Humidity.append(humidity_val)
            Dew_Point.append(dew_val)
            Visibility.append(visibility_val)
            Precipitation_Prob.append(precip_val)
            Rain.append(rain_val)
            Snow.append(snow_val)
            
            if len(City) >= max_rows:
                break
        if len(City) >= max_rows:
            break
    
    if len(City) >= max_rows:
        break
    
    time.sleep(1)        
```

```python
data = ({
    "City": City,
    "Date": Date,
    "Hour": Hour,
    "Temperature": Temperature,
    "Feels Like": Feels_Like,
    "Wind Speed": Wind_Speed,
    "Wind Direction": Wind_Direction,
    "Humidity": Humidity,
    "Dew Point": Dew_Point,
    "Visibility": Visibility,
    "Probability of Precipitation": Precipitation_Prob,
    "Rain": Rain,
    "Snow": Snow
})

df = pd.DataFrame(data)
df
```
